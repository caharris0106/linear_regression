```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('mlr03.xls')
df
```

```python
Y = df['FINAL']
df['ones'] = 1
X = df[['ones', 'EXAM1', 'EXAM2', 'EXAM3']]

w = np.linalg.solve( X.T.dot(X), X.T.dot(Y) )
Yhat = X.dot(w)

d1 = Y - Yhat
d2 = Y - Y.mean()
r2 = 1 - d1.dot(d1)/d2.dot(d2)
print(r2)
```
